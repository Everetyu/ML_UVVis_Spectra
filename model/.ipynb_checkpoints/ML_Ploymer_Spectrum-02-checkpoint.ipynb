```python
X_train = data.drop(["Abs"],axis=1)
y_train = data["Abs"]
print(type(X_train))
```

```python
# 训练集
X_train = data.drop(["Abs"],axis=1)
y_train = data["Abs"]

# 测试集
data = pd.read_excel("data_all_test.xlsx")
X_test = data.drop(["Abs"],axis=1)
y_test = data["Abs"]
```

```python
# 线性回归LinearRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error

LR = LinearRegression()
LR.fit (X_train, y_train)

# 模型的预测
LR_pred_train = LR.predict(X_train)
LR_pred_test = LR.predict(X_test)

pearson_train = np.corrcoef(y_train, LR_pred_train)
pearson_test = np.corrcoef(y_test, LR_pred_test)
MAPE = mean_absolute_percentage_error(y_test, LR_pred_test)
RMSE = np.sqrt(mean_squared_error(y_test, LR_pred_test))

print("线性回归的默认评估值为：", LR.score(X_test, y_test))
print("Pearson Coeffient train:",pearson_train)  #皮尔逊相关系数可以使用corrcoef()Numpy中的方法计算
print("Pearson Coeffient test:",pearson_test)
print("MAPE:", MAPE)
print("RMSE:", RMSE)
```

LinearRegression(fit_intercept=True,normalize=False,copy_X=True,n_jobs=1)

fit_intercept：bool类型，默认为True，表示是否计算截距(即y=wx+b中的b)，推荐设置为True。

normalize：bool类型，默认为False，表示是否对各个特征进行标准化默认方法是：减去均值并除以l2范数），推荐设置为True。如果设置为False，则建议在输入模型之前，手动进行标准化。标准化的好处：加速收敛、提升精度。

n_jobs：数值型，默认值为None。表示使用多少个处理器进行运算，当模型有多个回归目标( _targets > 1 )时，加速效果比较明显。推荐设置为 -1 ，此时会调用所有处理器。


```python
# KNN回归模型
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error

KNNR = KNeighborsRegressor(5)
KNNR.fit (X_train, y_train)

# 模型的预测
KNNR_pred_train = KNNR.predict(X_train)
KNNR_pred_test = KNNR.predict(X_test)

pearson_train = np.corrcoef(y_train, KNNR_pred_train)
pearson_test = np.corrcoef(y_test, KNNR_pred_test)

MAPE_train = mean_absolute_percentage_error(y_train, KNNR_pred_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, KNNR_pred_train))
MAE_train = mean_absolute_error(y_train, KNNR_pred_train)


MAPE_test = mean_absolute_percentage_error(y_test, KNNR_pred_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, KNNR_pred_test))
MAE_test = mean_absolute_error(y_test, KNNR_pred_test)

print("KNN回归的默认评估值为：", KNNR.score(X_test, y_test))
print("Pearson Coeffient train:",pearson_train)
print("Pearson Coeffient test:",pearson_test)
print("MAPE_train:", MAPE_train)
print("RMSE_train:", RMSE_train)
print("MAE_train:", MAE_train)
print("MAPE_test:", MAPE_test)
print("RMSE_test:", RMSE_test)
print("MAE_test:", MAE_test)
```

- n_neighbors：这个值就是指 KNN 中的 “K”了。前面说到过，通过调整 K 值，算法会有不同的效果。
- weights（权重）：最普遍的 KNN 算法无论距离如何，权重都一样，但有时候我们想搞点特殊化，比如距离更近的点让它更加重要。这时候就需要 weight 这个参数了，这个参数有三个可选参数的值，决定了如何分配权重。参数选项如下：
        • 'uniform'：不管远近权重都一样，就是最普通的 KNN 算法的形式。
        • 'distance'：权重和距离成反比，距离预测目标越近具有越高的权重。
        • 自定义函数：自定义一个函数，根据输入的坐标值返回对应的权重，达到自定义权重的目的。
- algorithm：在 sklearn 中，要构建 KNN 模型有三种构建方式，1. 暴力法，就是直接计算距离存储比较的那种放松。2. 使用 kd 树构建 KNN 模型 3. 使用球树构建。 其中暴力法适合数据较小的方式，否则效率会比较低。如果数据量比较大一般会选择用 KD 树构建 KNN 模型，而当 KD 树也比较慢的时候，则可以试试球树来构建 KNN。参数选项如下：
		• 'brute' ：蛮力实现
		• 'kd_tree'：KD 树实现 KNN
		• 'ball_tree'：球树实现 KNN 
		• 'auto'： 默认参数，自动选择合适的方法构建模型
不过当数据较小或比较稀疏时，无论选择哪个最后都会使用 'brute'
		
- leaf_size：如果是选择蛮力实现，那么这个值是可以忽略的，当使用KD树或球树，它就是是停止建子树的叶子节点数量的阈值。默认30，但如果数据量增多这个参数需要增大，否则速度过慢不说，还容易过拟合。
- p：和metric结合使用的，当metric参数是"minkowski"的时候，p=1为曼哈顿距离， p=2为欧式距离。默认为p=2。
- metric：指定距离度量方法，一般都是使用欧式距离。
		• 'euclidean' ：欧式距离
		• 'manhattan'：曼哈顿距离
		• 'chebyshev'：切比雪夫距离
		• 'minkowski'： 闵可夫斯基距离，默认参数
- n_jobs：指定多少个CPU进行运算，默认是-1，也就是全部都算。

C: 惩罚系数，C值越大，则容忍错误程度越低，这意味着模型将更加拟合训练数据。

kernel: 核函数类型，通常有线性核函数（linear）、多项式核函数（polynomial）、径向基函数（rbf）等。

gamma: 核函数宽度参数，用于控制选择的支持向量的数量。

degree: 多项式核函数的度数，通常只有使用多项式核函数时才会用到。

coef0: 某些核函数中的常量项。

tol: 求解器中的容差值。

max_iter: 求解器的最大迭代次数。

```python
# 岭回归模型
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error

#正则化系数alpha=0.1
Rid = Ridge(alpha=1)
Rid.fit (X_train, y_train)

# 模型的预测
Rid_pred_train = Rid.predict(X_train)
Rid_pred_test = Rid.predict(X_test)

pearson_train = np.corrcoef(y_train, Rid_pred_train)
pearson_test = np.corrcoef(y_test, Rid_pred_test)

MAPE_train = mean_absolute_percentage_error(y_train, Rid_pred_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, Rid_pred_train))
MAE_train = mean_absolute_error(y_train, Rid_pred_train)

MAPE_test = mean_absolute_percentage_error(y_test, Rid_pred_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, Rid_pred_test))
MAE_test = mean_absolute_error(y_train, Rid_pred_train)

print("岭回归的默认评估值为：", Rid.score(X_test, y_test))
print("Pearson Coeffient train:",pearson_train)
print("Pearson Coeffient test:",pearson_test)
print("MAPE_train:", MAPE_train)
print("RMSE_train:", RMSE_train)
print("MAE_train:", MAPE_train)
print("MAPE_test:", MAPE_test)
print("RMSE_test:", RMSE_test)
print("MAE_test:", MAPE_test)
```

岭回归（Ridge Regression），也称为Tikhonov正则化，是一种专门用于共线性数据分析的线性回归方法。它通过在损失函数中添加一个正则化项来解决标准最小二乘法的一些问题。这种正则化可以减少模型的复杂度，从而防止过拟合，同时仍然保留模型在数据上的拟合能力。岭回归在scikit-learn中通过Ridge类实现。以下是一些岭回归中重要的参数：

alpha：这是正则化强度的控制参数。alpha的值越大，正则化强度越高，这意味着减小模型的复杂度但可能导致对训练数据的拟合程度下降。相反，如果alpha的值太小，模型可能会过拟合训练数据。选择一个合适的alpha值通常是通过交叉验证来完成的。

fit_intercept：布尔值，指定是否计算该模型的截距。如果设置为False，则认为数据已经居中（即均值为零），并且模型中不会包含截距项。默认值为True。

normalize：布尔值，如果为True，则在回归之前，输入数据(X)会被减去均值并除以L2范数。如果fit_intercept设置为False，则此参数将被忽略。默认值为False。从scikit-learn 0.24版本开始，建议使用StandardScaler进行预处理。

copy_X：布尔值，如果为True，则对输入X进行复制，否则可能会被覆盖。默认值为True。

max_iter：最大迭代次数。对于特定求解器，这代表了达到收敛的最大迭代次数。

solver：用于计算岭回归系数的求解器。可选的求解器有auto，svd，cholesky，lsqr，sparse_cg，sag，和saga。默认值是auto，意味着根据数据类型自动选择求解器。

tol：求解的精度设定。对于指定的求解器，该参数控制求解的精度。

选择这些参数时，通常需要根据具体问题和数据集的特性进行调整。例如，alpha的选择通常依赖于交叉验证来确定最优值。而求解器的选择可能会根据数据的大小和特性来决定使用最合适的算法。

```python
# LASSO回归模型
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error

Las = Lasso()
Las.fit (X_train, y_train)

# 模型的预测
Las_pred_train = Las.predict(X_train)
Las_pred_test = Las.predict(X_test)

pearson_train = np.corrcoef(y_train, Las_pred_train)
pearson_test = np.corrcoef(y_test, Las_pred_test)
MAPE = mean_absolute_percentage_error(y_test, Las_pred_test)
RMSE = np.sqrt(mean_squared_error(y_test, Las_pred_test))

print("LASSO回归的默认评估值为：", Las.score(X_test, y_test))
print("Pearson Coeffient train:",pearson_train)
print("Pearson Coeffient test:",pearson_test)
print("MAPE:", MAPE)
print("RMSE:", RMSE)
```

LASSO回归模型是通过Lasso类实现的，它是线性模型的一种，通过在损失函数中添加L1范数作为正则项来进行约束，以此实现特征的稀疏选择，这有助于在预测模型中去除不重要的特征，从而提高模型的泛化能力。

Lasso类的重要参数如下：

alpha : float, 默认值=1.0

正则化强度；必须是一个正浮点数。alpha越大，正则化效果越强，模型越倾向于选择更少的特征。

fit_intercept : bool, 默认值=True

指定是否需要计算该模型的截距。如果设置为False，则假定数据已经居中。

normalize : bool, 默认值=False（在0.24版本之后这个参数将会被移除）

当fit_intercept设置为True时，如果normalize也被设置为True，则回归X将会在回归前被标准化，通过减去均值并除以l2范数。如果你想要一个标准化的解决方案，建议使用sklearn.pipeline.Pipeline中的StandardScaler。

precompute : bool or array-like, 默认值=False

是否使用预先计算的Gram矩阵来加速计算。如果设置为'auto'，则会自动决定。Gram矩阵也可以作为参数传入。

max_iter : int, 默认值=1000

坐标下降求解器的最大迭代次数。

tol : float, 默认值=1e-4

优化的容忍度：如果优化更新小于tol，则优化代码检查收敛并停止。

warm_start : bool, 默认值=False

当设置为True时，重用前一次调用的解作为初始化，否则，就擦除前一个解。

positive : bool, 默认值=False

当设置为True时，强制系数为正值。

selection : string, 默认值='cyclic'

如果设置为'random'，则随机选择一个特征来更新。设置为'cyclic'使用循环遍历所有特征的顺序。'random'通常可以导致收敛的速度更快，尤其是当tol较高时。

random_state : int, RandomState instance or None, 默认值=None

在selection设置为'random'时，用于随机数生成器的种子。仅在selection策略为'random'时使用。

这些参数使得LASSO模型相当灵活，可以适应不同的数据集和需求。在实践中，最常调整的参数是alpha，因为它直接控制了模型的复杂度和所选择特征的数量。调整这些参数时，建议使用交叉验证来评估不同参数设置下模型的性能，例如，可以使用sklearn中的LassoCV类，它可以自动地选择最佳的alpha值。

多层感知机（MLP）是一种前馈人工神经网络模型，它可以用于分类或回归问题。在回归问题中，多层感知机模型试图预测连续值的输出。MLP由多个层组成，包括一个输入层、一个或多个隐藏层，以及一个输出层。以下是构建MLP回归模型时需要考虑的一些主要参数：

1. 网络结构参数

隐藏层的数量和大小 (hidden_layer_sizes): 这是一个用于定义隐藏层的数量及其神经元（节点）数量的参数。例如，(100,) 表示有一个隐藏层，该层有100个神经元。

激活函数 (activation): 决定了层中神经元的输出方式。常见的激活函数包括 relu（线性整流函数）、tanh（双曲正切函数）、logistic（逻辑sigmoid函数）。激活函数对模型的能力和收敛速度有重要影响。

2. 训练算法参数

求解器 (solver): 用于权重优化的算法。常见的求解器包括 adam（适用于大规模数据或参数的情况）、sgd（随机梯度下降）、lbfgs（适用于小型数据集的拟牛顿方法）。不同的求解器适用于不同的数据集和模型复杂性。

学习率 (learning_rate): 在用随机梯度下降训练模型时，学习率控制了权重调整的幅度。有些求解器（如sgd）允许设置学习率，而其他求解器（如adam和lbfgs）则有固定的学习率策略。

学习率调整策略 (learning_rate_init, learning_rate): 对于sgd求解器，可以设置初始学习率（learning_rate_init）和学习率调整策略（learning_rate），如固定学习率、逐渐衰减的学习率等。

3. 正则化和停止条件

正则化项 (alpha): L2正则化项的系数，用于惩罚过于复杂的模型，防止过拟合。

最大迭代次数 (max_iter): 训练算法的最大迭代次数。达到最大迭代次数时，训练将停止，即使模型没有收敛。

容忍度 (tol): 用于优化的容忍度，当模型的损失值在连续几次迭代中改善小于此值时，算法将停止训练。

4. 其他参数

随机状态 (random_state): 控制权重的随机初始化方式，保证结果的可重现性。

批量大小 (batch_size): 在sgd和adam求解器中使用，表示每次迭代中使用的样本数量。

选择适当的参数对模型性能有很大影响。通常，需要通过交叉验证和实验来找到最佳参数组合。模型的调优是一个迭代过程，需要根据特定问题的性质和数据集的特点进行调整。

```python
# 使用随机森林的算法
from sklearn import ensemble
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error
# from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# 构建随机森林
RFR = ensemble.RandomForestRegressor(n_estimators=60) #, random_state=1234
# 随机森林的拟合
# print(y_train.shape)
RFR.fit(X_train, y_train)

# 模型的预测
RFR_pred_train = RFR.predict(X_train)
RFR_pred_test = RFR.predict(X_test)

pearson_train = np.corrcoef(y_train, RFR_pred_train)
pearson_test = np.corrcoef(y_test, RFR_pred_test)

MAPE_train = mean_absolute_percentage_error(y_train, RFR_pred_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, RFR_pred_train))
MAE_train = mean_absolute_error(y_train, RFR_pred_train)

MAPE_test = mean_absolute_percentage_error(y_test, RFR_pred_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, RFR_pred_test))
MAE_test = mean_absolute_error(y_test, RFR_pred_test)

print("随机森林回归的默认评估值为：", RFR.score(X_test, y_test))
print("Pearson Coeffient train:",pearson_train)
print("Pearson Coeffient test:",pearson_test)
print("MAPE_train:", MAPE_train)
print("RMSE_train:", RMSE_train)
print("MAE_train:", MAE_train)
print("MAPE_test:", MAPE_test)
print("RMSE_test:", RMSE_test)
print("MAE_test:", MAE_test)
```

在scikit-learn库中，随机森林模型是通过RandomForestClassifier类（对于分类问题）和RandomForestRegressor类（对于回归问题）实现的。这两个类共享许多相同的参数，因为随机森林是通过集成多个决策树来工作的。以下是这些模型的一些主要参数：

1. n_estimators:
类型: int，默认值=100
描述: 森林里的树木数量。更多的树增加了模型的性能，但也会使模型的训练时间更长。
2. criterion:
对于RandomForestClassifier，默认值是gini，另一个可选的值是entropy。

类型: int，默认值=None

类型: int、RandomState实例或None，默认值=None
描述: 控制组件的随机性。

类型: int，默认值=0
描述: 在拟合和预测时控制详细程度。

类型: bool，默认值=False

在 scikit-learn 中，梯度提升决策树（Gradient Boosting Decision Trees，简称GBDT）是通过 GradientBoostingClassifier 类（用于分类问题）和 GradientBoostingRegressor 类（用于回归问题）实现的。GBDT 是一个强大的机器学习算法，用于解决分类和回归问题，通过逐步添加预测弱树模型，尝试减小模型误差，每一棵新添加的树都尝试修正前面所有树的预测残差。以下是一些主要的参数：

1. loss:
对于分类：默认值是deviance，表示对数似然损失函数。另外，对于二分类问题还可以选择exponential损失，后者会使模型表现得像AdaBoost。
对于回归：默认值是ls（最小二乘法），表示最小化均方误差。其他选项包括lad（最小绝对偏差）、huber（结合了ls和lad的优点）和quantile（用于执行分位数回归）。
2. learning_rate:
类型: float，默认值=0.1
描述: 每棵树对最终结果的贡献，即学习率。较低的值使模型更稳健，通常需要更多的树来训练，以达到相同的精度。
3. n_estimators:
类型: int，默认值=100
描述: 进行梯度提升的树的数量。更多的树会使模型更复杂，也容易导致过拟合。
4. subsample:
类型: float，取值在(0,1]，默认值=1.0
描述: 用于拟合各个基学习器的样本比例。如果小于1.0，则将采用随机梯度提升。这种策略可以减少过拟合。
5. criterion:
类型: string，仅限回归，默认值=friedman_mse。
描述: 用于测量分裂质量的函数。除了friedman_mse，还可以使用mse（均方误差）和mae（平均绝对误差）。
6. min_samples_split:
类型: int 或 float，默认值=2
描述: 分裂内部节点所需的最少样本数。
7. min_samples_leaf:
类型: int 或 float，默认值=1
描述: 在叶节点处需要的最小样本数。
8. min_weight_fraction_leaf:
类型: float，默认值=0.0
描述: 在所有叶节点处的权重总和中的最小加权分数。
9. max_depth:
类型: int，默认值=3
描述: 各个回归估计器的最大深度。限制树的深度可以有效防止过拟合。
10. min_impurity_decrease:
类型: float，默认值=0.0
描述: 如果节点的分裂导致不纯度的减少大于或等于这个值，则节点将会分裂。
11. max_features:
类型: int、float、string 或 None，默认值=None
描述: 查找最佳分割时要考虑的特征数量。这能够提供过拟合的控制，尤其是当特征数量很大时。
12. random_state:
类型: int、RandomState 实例或 None，默认值=None
描述: 控制随机数生成器的种子。这个参数是随机化的一部分，但并不控制迭代的种子或其他一些随机性。
13. max_leaf_nodes:
类型: int 或 None，默认值=None
描述: 以最佳优先级和最大叶子数来生长树。有助于控制模型的复杂度。
14. warm_start:
类型: bool，默认值=False
描述: 设置为True时，重用上一个调用的解决方案来适应并向集成中添加更多的估计器。
GBDT 是一个非常灵活且强大的模型，可以通过调整这些参数来优化性能。不过，适当的参数调整需要基于对数据的理解和多次实验。通常，需要在模型复杂度和泛化能力之间进行权衡，避免过拟合。

XGBoost（eXtreme Gradient Boosting）是一种高效且灵活的梯度提升框架，尽管它不是scikit-learn的一部分，但它与scikit-learn具有良好的兼容性，并且可以通过pip进行安装。XGBoost在很多方面优化了传统的GBDT，包括速度和性能，同时在处理稀疏数据上也有很好的表现。以下是一些XGBoost模型的主要参数，这些参数可以在初始化模型时设置，如xgboost.XGBClassifier()或xgboost.XGBRegressor()。

常用的XGBoost参数可以分为几个大类：
1. 通用参数（General Parameters）：
booster: 选择每次迭代的模型，有gbtree（树模型）、gblinear（线性模型）和dart（Dropouts meet Multiple Additive Regression Trees）。
nthread: 并行线程的数量。默认情况下，它设置为最大可能的线程数。
2. Booster参数：
eta (learning_rate in sklearn): 通过减少每一步的权重，防止过拟合。范围为[0,1]。
min_child_weight: 决定最小叶子节点样本权重和，用于控制过拟合。
max_depth: 树的最大深度，用于控制过拟合。
max_leaf_nodes: 最大叶子节点数。
gamma: 在树的叶子节点做进一步分割所需的最小损失减少量。增加这个值会使模型更保守。
subsample: 控制每棵树随机采样的比例，减少这个参数的值，算法会更加保守，防止过拟合。
colsample_bytree: 每棵树的列的子样本率。
3. 学习任务参数（Learning Task Parameters）：
objective: 指定学习任务和相应的学习目标，比如binary:logistic用于二分类的逻辑回归，reg:squarederror用于回归任务的均方误差。
eval_metric: 评估模型性能的指标，如rmse用于回归，logloss用于二分类，merror用于多分类等。
seed: 随机数种子，用于产生可复现的结果。
4. 命令行参数（Command Line Parameters）（仅在通过命令行界面使用XGBoost时相关）。
注意事项：
实际使用中，eta、gamma、max_depth、min_child_weight、subsample和colsample_bytree是最常调的参数。
参数的最佳组合通常通过交叉验证来确定。
XGBoost还提供了早停（early stopping）的功能，可以在验证集上的表现不再提升时自动停止训练，这在实际应用中非常有用，可以省去手动调参的麻烦。
XGBoost的灵活性和效率使其成为处理各种机器学习任务的流行选择。正确调参是发挥其最大性能的关键。

LightGBM是一个基于梯度提升框架的高效、灵活且可扩展的机器学习算法，由微软提供。它以其训练速度快、效率高、占用内存少而受到广泛欢迎。LightGBM使用基于直方图的算法，它将连续的特征值分桶（bin）到离散的bin中，从而减少计算量。此外，它还支持类别特征的直接输入，无需进行独热编码。

LightGBM的主要参数可以分为以下几个类别：

1. 核心参数（Core Parameters）

task: 指定任务类型，比如train、predict等。

boosting_type: 指定提升类型，比如gbdt（传统的梯度提升决策树），rf（随机森林），dart（Dropouts meet Multiple Additive Regression Trees），goss（Gradient-based One-Side Sampling）。

objective: 指定目标函数，比如binary用于二分类任务，regression用于回归任务等。

metric: 指定评估指标，比如binary_logloss、auc用于二分类，l1、l2用于回归等。

2. 学习控制参数（Learning Control Parameters）

learning_rate: 学习速率，较小的值意味着需要更多的树来进行训练。

num_leaves: 一棵树上的叶节点数，过多会增加过拟合的风险。

tree_learner: 用于并行学习，serial表示单台机器训练，feature、data、voting分别对应不同的并行方式。

max_depth: 树的最大深度，限制树模型的复杂度，有利于防止过拟合。

min_data_in_leaf: 叶子可能具有的最小记录数，过小会导致过拟合。

bagging_fraction: 不进行重采样的情况下随机选择部分数据可以加快训练速度。

feature_fraction: 在进行训练之前选择部分特征可以防止过拟合。

lambda_l1, lambda_l2: L1和L2正则化项，用于防止过拟合。

3. IO参数（IO Parameters）

max_bin: 最大的bin数量，较小的值可以提高运行速度，同时也可能减少模型精度。

categorical_feature: 指定类别特征，不需要转换成独热编码。

4. 其他参数

还有一些其他的特定参数，用于调整模型的保存、加载、日志记录等细节。

在使用LightGBM时，合理调整这些参数对于提高模型性能、减少模型训练时间和防止过拟合是至关重要的。建议通过交叉验证等方式来寻找最优的参数组合。

```python
#使用CatBoost
import catboost
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, mean_absolute_error

#CatBoost回归
CatBoost_R =  catboost.CatBoostRegressor(iterations=500, learning_rate=0.01, loss_function='RMSE',verbose=False)

CatBoost_R.fit(X_train, y_train)

# 模型的预测
CatBoost_R_pred_train = CatBoost_R.predict(X_train)
CatBoost_R_pred_test = CatBoost_R.predict(X_test)

pearson_train = np.corrcoef(y_train, CatBoost_R_pred_train)
pearson_test = np.corrcoef(y_test, CatBoost_R_pred_test)

MAPE_train = mean_absolute_percentage_error(y_train, CatBoost_R_pred_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, CatBoost_R_pred_train))
MAE_train = mean_absolute_error(y_train, CatBoost_R_pred_train)

MAPE_test = mean_absolute_percentage_error(y_test, CatBoost_R_pred_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, CatBoost_R_pred_test))
MAE_test = mean_absolute_error(y_test, CatBoost_R_pred_test)

print("CatBoost回归的默认评估值为：", CatBoost_R.score(X_test, y_test))
print("Pearson Coeffient train:",pearson_train)
print("Pearson Coeffient test:",pearson_test)
print("MAPE_train:", MAPE_train)
print("RMSE_train:", RMSE_train)
print("MAE_train:", MAE_train)
print("MAPE_test:", MAPE_test)
print("RMSE_test:", RMSE_test)
print("MAE_test:", MAE_test)
```

CatBoost是一个高性能的开源库，用于梯度提升分类和回归问题。由于其对类别特征的原生支持和处理能力，它在各种机器学习任务中特别受欢迎。
CatBoost具有减少模型过拟合的能力，同时提供高精度的模型，且易于使用。以下是CatBoost模型的一些主要参数：

1. 核心参数

iterations (或 num_boost_round): 提升树的数量。更多的树可以提高模型的准确度，但同时也可能导致过拟合。

learning_rate (或 eta): 学习速率，用于减少每一步的权重，以防止过拟合。较小的值通常与更多的树一起使用，可以提高模型的质量。

depth: 树的深度，用于控制树的复杂度。较大的深度可以提高模型的准确度，但也会增加计算成本和过拟合的风险。

l2_leaf_reg (正则化项): L2正则化项的系数。增加此参数的值将使模型更保守，有助于防止过拟合。

2. 类别特征处理参数

cat_features: 指定输入数据中的类别特征列。CatBoost会自动将这些类别特征处理成数值，无需进行预处理。

one_hot_max_size: 类别特征的独热编码最大值。如果类别特征的唯一值数量不超过此参数，则使用独热编码。

3. 训练控制参数

bootstrap_type: 用于控制采样的方法。常见的选项包括Bayesian、Bernoulli、MVS等。

subsample: 在每次迭代中用于训练的数据样本比例（仅当使用Poisson或Bernoulli作为bootstrap_type时有效）。

random_strength: 每个分裂的分数的随机性。增加这个参数的值会使模型更加保守。

early_stopping_rounds: 如果在指定的迭代轮数内，验证集的评分没有改进，则停止训练。这有助于避免过拟合并减少训练时间。

4. 其他参数

eval_metric: 用于训练过程中的模型评估的度量指标。

loss_function: 指定模型的损失函数，这取决于具体的任务类型（例如Logloss、CrossEntropy、RMSE等）。

use_best_model: 如果设置为True，则使用在验证集上表现最好的模型进行预测。

CatBoost的参数非常灵活，可以根据具体的数据集和任务需求进行调整。合理的参数调整可以大大提高模型的效果和性能。为了获得最佳的模型参数，建议使用CatBoost提供的参数调优功能，如网格搜索（Grid Search）或随机搜索（Random Search）。

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer
from scipy.stats import pearsonr

# 所有模型列表
model = [LR,
         KNNR,
         SVMR,
         Rid,
         Las,
         MLPR,
         RFR,
         ExtraTree_R,
         GBDT_R,
         XGBoost_R,
         LightGBM_R,
         CatBoost_R]

# 自定义评分函数，它将返回皮尔逊相关系数
def pearson_correlation(y_true, y_pred):
    return pearsonr(y_true, y_pred)[0]  # 只返回相关系数

# 创建一个scorer对象
pearson_scorer = make_scorer(pearson_correlation)

# 假设已经定义了一个pipeline对象和训练数据X_train, y_train

for M in model:
    #创建一个管道（Pipeline）实例,包含标准化方法和模型算法
    pipeline = make_pipeline(StandardScaler(), M)

    # 设置交叉验证折数cv=10 表示使用带有十折的StratifiedKFold，再把管道和数据集传到交叉验证对象中
    scores = cross_val_score(pipeline, X_train, y_train, scoring=pearson_scorer, cv=10, n_jobs=1)
    
    print("model:",M)
    print('Cross Validation scores: %s' % scores)
    print('Cross Validation: %.8f +/- %.8f' % (np.mean(scores),np.std(scores)))
```
